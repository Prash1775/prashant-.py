```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("SAP-4000.csv")
```

```python
# 1. Clean column names
df.columns = df.columns.str.strip()

# 2. Basic Descriptive Analysis
print("\nBasic Info:\n", df.info())
print("\nSummary Stats:\n", df.describe())
print("\nValue Counts:\n", df['Region'].value_counts())
print("\nParent Education Levels:\n", df['Parent Education'].value_counts())
```

```python
# 3. Pie Chart â€“ Gender Distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Gender Distribution")
plt.tight_layout()
plt.show()
```

```python
# Gender-based performance
gender_avg = df.groupby('Gender')['Exam_Score'].mean()
print("Average Exam Score by Gender:")
print(gender_avg)
```

```python
# 4. Bar Chart â€“ Average Exam Score by Region
avg_score_region = df.groupby('Region')['Exam_Score'].mean()
plt.figure(figsize=(6, 5))
avg_score_region.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Average Exam Score by Region")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Region-based performance
region_avg = df.groupby('Region')['Exam_Score'].mean()
print("\nAverage Exam Score by Region:")
print(region_avg)
```

```python
# 5. Line Chart â€“ Average Exam Score by Study Hours (grouped)
df['Study_Hour_Group'] = pd.cut(df['HoursStudied/Week'], bins=[0, 5, 10, 15, 20, 25], labels=['0-5', '6-10', '11-15', '16-20', '21-25'])
avg_score_by_hours = df.groupby('Study_Hour_Group')['Exam_Score'].mean()
plt.figure(figsize=(7, 5))
avg_score_by_hours.plot(kind='line', marker='o', color='green')
plt.title("Exam Score by Study Hours Group")
plt.xlabel("Study Hours per Week")
plt.ylabel("Average Exam Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 6. Line Graph â€“ Attendance vs Exam Score (trend)
attendance_score = df.sort_values('Attendance(%)')[['Attendance(%)', 'Exam_Score']]
plt.figure(figsize=(8, 5))
plt.plot(attendance_score['Attendance(%)'], attendance_score['Exam_Score'], linestyle='-', marker='', alpha=0.3)
plt.title("Exam Score Trend Based on Attendance")
plt.xlabel("Attendance (%)")
plt.ylabel("Exam Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Correlation insights
print("\nCorrelation Matrix:")
print(df[['HoursStudied/Week', 'Attendance(%)', 'Exam_Score']].corr())
```

```python



# Parent Education impact
parent_edu_avg = df.groupby('Parent Education')['Exam_Score'].mean()
print("\nAverage Exam Score by Parent Education:")
print(parent_edu_avg)


# Tutoring vs No Tutoring
tutoring_avg = df.groupby('Tutoring')['Exam_Score'].mean()
print("\nAverage Exam Score with/without Tutoring:")
print(tutoring_avg)

# Study hour group comparison
print("\nAverage Score by Study Hour Groups:")
print(avg_score_by_hours)
```
